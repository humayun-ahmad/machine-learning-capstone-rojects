```python
def decompose_category(x):
    if x == "England":
        return 'England'
    elif x == "Argentina":
        return 'Argentina'
    elif x == "France":
        return 'France'
    elif x == "Spain":
        return 'Spain'
    elif x == "Brazil":
        return 'Brazil'
    else:
        return 'Others'
    

data['Nationality'] = data['Nationality'].apply(decompose_category)
data.Nationality.value_counts()
```
