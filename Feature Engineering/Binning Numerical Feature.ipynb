```python
def age_range(x):
    if x>17 and x<30:
        return '17-30'
    elif x>30 and x<40:
        return '30-40'
    elif x>40 and x<50:
        return '40-50'
    else:
        return '50-60'
    
data['Age'] = data['Age'].apply(age_range)
data['Age'].value_counts()
# data
```

```python
data['DistanceFromHome'].value_counts()
```

```python
def distane_from_home(data):
    if data <= 2:
        return 'Near'
    elif data > 2 and data < 5:
        return 'Far'
    elif data > 5:
        return 'Extremely Far'
    
data['DistanceFromHome'] = data['DistanceFromHome'].apply(distane_from_home)
data['DistanceFromHome'].value_counts()
        
```

```python
data['TotalWorkingYears'].value_counts()
```

```python
def experiment_category(data):
    if data < 2:
        return 'novice'
    elif data > 2 and data < 5:
        return 'intermediate'
    elif data > 5:
        return 'experienced'

# Apply the function to the 'TotalWorkingYears' column
data['TotalWorkingYears'] = data['TotalWorkingYears'].apply(experiment_category)

# Get value counts
print(data['TotalWorkingYears'].value_counts())
```
